```python
%matplotlib inline
import numpy as np
import random as rand
import matplotlib.pyplot as plt
from random import sample
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
x1 = np.linspace(-100,100,1000)
x2 = np.linspace(-100,100,1000)
x1, x2 = np.meshgrid(x1, x2)

def function1(x1, x2, f):
    if f==1:
        f1 = (1 - x1)**2 + 100*(x2 - x1**2)**2
    else:   
        f1 = (np.sin(3*x1)*np.cos(3*x2))/((x1**2)*(x2**2)+1)
    return f1
f=1
f1 = function1(x1, x2, f)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax3d = plt.axes(projection='3d')
ax3d.plot_surface(x1, x2, f1,cmap='plasma')
ax3d.set_title('Surface Plot in Matplotlib')
ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z')

plt.show()
```

```python

generacion=0
anterior=0
generaciones=[]
generaciones_best=[]
fitnes=[]
fitnes_best=[]
promediop=[]
fitn=0

for f in funciones:
    for algoritmo in algoritmos:
        for N in Nl:
            for F in Fl:
                for C in Cl:
                    poblacion = np.random.random((N,n))*(xmax - xmin) + xmin
                                        
                    promediop=[]
                    generaciones=[]  
                    fitnes=[]
                    generacion=0
                    while True:

                        generacion=generacion+1
                        
                        for individuo in range(len(poblacion)):
                            
                            # Se escogen 3 individuos no repetidos de la poblacion 
                            individuos = sample(range(N), 3)
                            xr = poblacion[individuos[0]]
                            xs = poblacion[individuos[1]]
                            xt = poblacion[individuos[2]]

                            xi = poblacion[individuo]
                            
                            # Se crea el descendiente provisional
                            if algoritmo==1:
                                xoff_p = xt + F*(xr - xs)
                            else:
                                for i in range(len(poblacion)):
                                    x = poblacion[i]
                                    z3= function1(x[0], x[1], f)
                                    if anterior>z3 or i==0:
                                        anterior=z3
                                        x_best=poblacion[i]
                                    #    z3_best=function1(x_best[0], x_best[1], f)
                                #fitnes_best.append(z3_best)    
                                #generaciones_best.append(generacion)

                                xoff_p = x_best + F*(xs - xt)
                                
                            # Se hace la recombinacion 
                            xoff = np.zeros(n)

                            for i in range(len(xoff_p)):
                                xoff[i] = xi[i] if np.random.random() < C else xoff_p[i]
                            # Se selecciona al mejor individuo 
                            z1 = function1(xoff[0], xoff[1], f)
                            z2 = function1(xi[0], xi[1], f)

                            if z1 < z2:
                                poblacion[individuo] = xoff
                            else:
                                poblacion[individuo] = xi

                            z = function1(poblacion[individuo, 0],poblacion[individuo, 1], f)




                            if z < 2:
                                print(poblacion[individuo], z1, z2, generacion)
                                break
                        #guardamos fitnes para cada generacin


                        fitn=0
                        for j in range(len(poblacion)):
                            cont1 = poblacion[j]
                            fit= function1(cont1[0], cont1[1], f)
                            fitn=fitn+fit
                        promedio=fitn/len(poblacion)
                        promediop.append(promedio)
                        generaciones.append(generacion)
                        fitnes.append(z)
                        if z < 2:
                                break


                    fig, ax = plt.subplots() # Create the figure and axes object

                    # Plot the first x and y axes:

                    
                    plt.plot(generaciones, fitnes) 
                    plt.plot(generaciones, promediop)   
                    plt.legend(["fitne Mejor", "fitnes average"])
                    plt.xlabel("generaciones ", size = 16,)
                    plt.ylabel("fitnes", size = 16)

                    plt.title(f' N={N} F={F} C={C} Algo={algoritmo} funcion={f}')
                    plt.savefig(f'{N}_F={F}_C={C} Algo={algoritmo} funcion={f}.png') 
                    plt.show()    


```
